<b> Plot data directly from a Pandas dataframe. </b>

```python
# Read in Gapminder Data
data = pandas.read_csv('gapminder_gdp_oceania.csv', index_col='country')

# Extract year from last 4 characters of each column name
years = data.columns.str.strip('gdpPercap_')
# Convert year values to integers, saving results back to dataframe
data.columns = years.astype(int)
# Plot
data.loc['Australia'].plot()
```

<b> Select and transform data, then plot it. </b>

```python
# Transform data on fly so years are rows and countries columns, and plot
data.T.plot()
plt.ylabel('GDP per capita')
```

<b> Test fancier barplot styles using ggplot </b>

```python
plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.ylabel('GDP per capita')
```

<b> Call matplotlib plot function directly </b>

```python
#Plot Australian GDP over time in years
years = data.columns
gdp_australia = data.loc['Australia']
# g-- specifies a green dashed line, 
plt.plot(years, gdp_australia, 'g--')
```

<b> Plot multiple datasets together </b>

```python
# Select two countries' worth of data.
gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']

# Plot with differently-colored markers.
plt.plot(years, gdp_australia, 'b-', label='Australia')
plt.plot(years, gdp_nz, 'g-', label='New Zealand')

# Create legend (uses labels specified earlier)
plt.legend(loc='lower right')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')
```

<b> Scatter plot correlating the GDP of Australia and New Zealand </b>

```python
plt.scatter(gdp_australia, gdp_nz)
plt.title("GDP")
plt.xlabel('Australia')
plt.ylabel('New ZeaLand')
```

<b> Minimum and maximum GDP per capita over time for all the countries in Europe. </b>

```python
data_europe = pandas.read_csv('data/gapminder_gdp_europe.csv', index_col='country')
# Minimum
data_europe.min().plot(label='min')
# Maximum
data_europe.max().plot(label='max')
# Formatting
plt.legend(loc='best')
plt.xticks(rotation=90)
```

<b> Scatter plot of relationship between minimum and maximum GDP per capita among the countries in Asia each year </b>

```python
data_asia = pandas.read_csv('data/gapminder_gdp_asia.csv', index_col='country')
data_asia.describe().T.plot(kind='scatter', x='min', y='max', title = "GDP in Asian Countries")
```

<b> Correlation between GDP and life expectancy for 2007, normalizing marker size by population: </b>

```python
data_all = pandas.read_csv('gapminder_all.csv', index_col='country')
# Scatter defines it to be a scatter plot
# x specifies the x axis data
# Y specifies the y axis data
# s specifies the size of the marker - in this cased based on the population
data_all.plot(kind='scatter', x='gdpPercap_2007', y='lifeExp_2007',
              s=data_all['pop_2007']/1e6)
```
